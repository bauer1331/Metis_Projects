```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from seaborn import plt
```

```python
dogs  = pd.read_csv('../../../local_files/Projects/Instagram_Likes_Local/data/dogs_alltime.csv')
```

```python
tobypuff = dogs[dogs['username']=='tobypuff'].likes
tobypuff = list(tobypuff)
len(tobypuff)
```

```python
posts = list(range(1,len(tobypuff)))
posts.sort(reverse=True)
tobypuff = list(zip(posts, tobypuff))
tobypuff = pd.DataFrame(tobypuff,columns=['posts', 'likes'])
```

```python
tmean = tobypuff.likes.mean()
tstd = np.std(tobypuff.likes)
print((tmean+ tstd*.3), (tmean - tstd*.5))
```

```python
plt.plot(tobypuff['posts'],tobypuff['likes'],label='Likes')
plt.axhline(y=(tmean), color='g', linestyle='-', label = 'Above Average Likes = 2785')
plt.axhline(y=(tmean - tstd*.5), color='r', linestyle='-', label = 'Below Average Likes = 2128')
plt.legend(loc=1)
plt.title('Tobypuff Total Likes over Time')
plt.xlabel("Posts")
plt.ylabel("Count of Likes")
```
